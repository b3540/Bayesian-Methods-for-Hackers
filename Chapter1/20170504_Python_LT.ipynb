# コイン投げ
表の出る確率がθである１枚のコインがある。このコインを投げたとき１回目→表、２回目→表、３回目→表、４回目→裏と出たとする。このとき「表のでる確率θ」の確率分布を求めよ。


## コイン投げ１回目

```python
xvar = np.arange(0,1,0.01)
```

```python
y = lambda x:x
yvar = y(xvar)
plt.plot(xvar,yvar,label="y=θ")
plt.plot(xvar,2*yvar,label="y=2θ")

plt.title("事後分布 $π_1(θ| $表$)=2θ$")

plt.legend(loc = "best")

plt.savefig("pi1")
```

## コイン投げ２回目

```python
y = lambda y :y**2
yvar = y(xvar)
plt.plot(xvar,2*yvar,label="$y=2θ^2$")
plt.plot(xvar,3*yvar,label="$y=3θ^2$")

plt.title("事後分布 $π_2(θ| $表$)=3θ^2$")
plt.legend(loc = "best")

plt.savefig("pi2")
```

## コイン投げ３回目

```python
y = lambda y :y**3
yvar = y(xvar)

plt.plot(xvar,3*yvar,label="$y=3θ^3$")
plt.plot(xvar,4*yvar,label="$y=4θ^3$")

plt.title("事後分布 $π_3(θ| $表$)=4θ^3$")
plt.legend(loc = "best")

plt.savefig("pi3")
```

## コイン投げ４回目

```python
y = lambda y :(1-y) * y**3
yvar = y(xvar)
plt.plot(xvar,4*yvar,label="$y=4θ^3(1-θ)$")
plt.plot(xvar,20*yvar,label="$y=20θ^3(1-θ)$")

plt.title("事後分布 $π_4(θ| $裏$)=20θ^3(1-θ)$")
plt.xticks([0.25,0.5,0.75],[0.25,0.5,0.75])
plt.legend(loc = "best")

plt.savefig("pi4")
```

# sympyを使用した積分の例

```python
import sympy as sym
z = sym.symbols("z") # zをsympyの変数にする
f1 = z #    関数z1を定義　
f2 = 4/(1+z**2) #    関数f2を定義
```

```python
sym.integrate(f1,(z,0,1))  #    f1を0,1の範囲で積分
```

```python
sym.integrate(f2,(z,0,1)) #     f4を0,1の範囲で積分
```

```python
θ = sym.symbols('θ')
f = (θ**3)*(1-θ)
sym.integrate(f,(θ,0,1)) #fを0,1の範囲で積分
```
